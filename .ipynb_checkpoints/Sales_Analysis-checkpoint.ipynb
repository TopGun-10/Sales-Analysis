```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
fin_df = pd.read_csv("fin_df.csv")
```

```python
fin_df.dropna()
```

```python
fin_df['Month'] = fin_df["Order Date"].str[0:2]
# Dropping NA values
fin_df = fin_df.dropna(how="all")

fin_df.head()
```

```python
fin_df.head()
```

```python
# Cleaning Data
er_row = fin_df[fin_df["Order Date"].str[0:2] == "Or"]
er_row.head()
fin_df = fin_df[fin_df["Order Date"].str[0:2] != "Or"]
```

```python
fin_df["Month"]  = fin_df["Month"].astype('int32')
fin_df["Quantity Ordered"]  = pd.to_numeric(fin_df["Quantity Ordered"])
fin_df["Price Each"]  = pd.to_numeric(fin_df["Price Each"])
fin_df.head()
```

#  Q.  Which month had highest sales this year ?


```python
fin_df.groupby("Month")
fin_df.head()
```

```python
# Creating a new column "Total Price"
fin_df["Sales"] = fin_df["Quantity Ordered"] * fin_df["Price Each"]
fin_df.head()
```

```python
monthly_sales = fin_df.groupby("Month").sum()
print(monthly_sales)
monthly_sales["Sales"]

# In output we can see 12 th Month i.e DEcember was the best month for sales
```

```python
# plotting months vs sales

months = range(1,13)
plt.bar(months,monthly_sales["Sales"])
plt.xticks(months)
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.show()
```

# Q. In which city sales were highest ?

```python

fin_df["City"] = fin_df["Purchase Address"].apply(lambda x : x.split(",")[1]) +" "+ fin_df["Purchase Address"].apply(lambda x : x.split(",")[2].split(" ")[1])

fin_df.head()
```

```python
# summing up total sales for each city
city_sales = fin_df.groupby("City").sum()
print(city_sales["Sales"])
```

```python
# Lets Plot
cities = [city for city,df in fin_df.groupby("City")]
plt.bar(cities,city_sales["Sales"])
plt.xticks(cities,rotation="vertical",size = 10)
plt.xlabel("cities")
plt.ylabel("Sales ($)")
plt.show()
```

# Q . At  which time of the day sales were more ?


```python


fin_df["Order Date"] = pd.to_datetime(fin_df["Order Date"])
fin_df.head()
```

```python
# Getting an "Hour" column 
fin_df["Hour"] = fin_df["Order Date"].dt.hour
fin_df["Minute"] = fin_df["Order Date"].dt.minute
fin_df.head()
```

```python

hours = [ hour for hour,df in fin_df.groupby("Hour")]
```

```python
#  print(fin_df.groupby(["Hour"]).count().sum())
plt.plot(hours , fin_df.groupby(["Hour"]).count() )
plt.xticks(hours)
plt.xlabel("Hour")
plt.ylabel("Sales ")
plt.grid()
plt.show()

# Clearly we can see at 19th hour sales are more .
```
